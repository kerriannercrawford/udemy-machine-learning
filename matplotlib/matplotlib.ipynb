## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Set up plot
figure, axes = plt.subplots(figsize=(10, 10)) # width, height

# 3. Plot data
axes.plot(x, y)

# 4. Customize plot
axes.set(title='Simple Plot', xlabel='x-axis', ylabel='yaxis')

# 5. Save and show (you save the whole figure)
figure.savefig('sample-plot.png')
```

## Bar graph

```python
# Create a plot from a dictionary
nut_butter_prices = { 'Almond butter': 10, 'Peanut Butter': 8, 'Cashew butter': 12 }
figure, axis = plt.subplots()
axis.bar(nut_butter_prices.keys(), nut_butter_prices.values())
axis.set(title='Nut butter prices', ylabel='Price');
```

## Horizontal Bar Graph

```python
figure, axis = plt.subplots()
axis.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

## Histogram

```python
x = np.random.randn(1000)
figure, axis = plt.subplots()
axis.hist(x);
```

## Two Options for SubPlots

### Option 1

```python
figure, ((axis1, axis2), (axis3, axis4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different axis
axis1.plot(x, x/2)
axis2.scatter(np.random.random(10), np.random.random(10))
axis3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
axis4.hist(np.random.randn(1000));
```

### Option 2

```python
figure, axis = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
# Plot to each different index
axis[0,0].plot(x, x/2)
axis[0, 1].scatter(np.random.random(10), np.random.random(10))
axis[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
axis[1, 1].hist(np.random.randn(1000));
```

### Which one should you use? pyplot vs matplotlib OO method?
- When plotting something quickly, ok to use pyplot method
- When plotting something more advanced, use the OO method

```python
heart_disease.head()
```

```python
# Look at patients over 50
over_50 = heart_disease[heart_disease['age'] > 50]
over_50
```

```python
# Pyplot method
over_50.plot(kind='scatter', x='age', y='chol', c='target');
```

```python
## OO Method - sort of
figure, axis = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=axis);

# axis.set_xlim([45, 100]);
```

```python
## True OO method
figure, axes = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = axes.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])

# Customize
axes.set(title='Heart Disease and Cholesterol Levels', xlabel='Age', ylabel='Cholesterol')

# Add a legend
axes.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
axes.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
over_50
```

```python
# Subplot of chol, age, thalach
figure, (axes_0, axes_1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))
```

```python
%matplotlib inline
# Add data to axes 0
scatter = axes_0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])

# Customize the data
axes_0.set(title='Heart Disease and Cholesterol Levels', xlabel='Age', ylabel='Cholesterol')

axes_0.legend(*scatter.legend_elements(), title='Target')
axes_0.axhline(y=over_50['chol'].mean(), linestyle='--')

# Add data to axes 1
scatter = axes_1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'])

# Customize the data
axes_1.set(title='Heart Disease and Max Heart Rate', xlabel='Age', ylabel='Max Heart Rate')

axes_1.legend(*scatter.legend_elements(), title='Target')
axes_1.axhline(y=over_50['thalach'].mean(), linestyle='--')

# Add a title to the figure
figure.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
figure
```
